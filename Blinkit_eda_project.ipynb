#### group by (its sort the values in perticular group)

- sum(): Calculates the sum of values within each group.
- mean(): Calculates the mean (average) of values within each group.
- count(): Counts the number of non-NA values within each group.
- min(): Finds the minimum value within each group.
- max(): Finds the maximum value within each group.
- agg(): Allows you to apply multiple aggregation functions at once.

### creating data frame

```python
# Method 1: Creating a DataFrame from a dictionary of lists/arrays
data = {
    'roll': [1, 2, 3, 4, 5],
    'name': ['A', 'B', 'C', 'D', 'E'],
    'weight': [10.1, 20.2, 30.3, 40.4, 50.5]
}
df1 = pd.DataFrame(data)
print("DataFrame from dictionary:")
print(df1)
```

```python
# Method 1: Creating a DataFrame from a dictionary of lists/arrays
data = {
    'col1': [1, 2, 3, 4, 5],
    'col2': ['A', 'B', 'C', 'D', 'E'],
    'col3': [10.1, 20.2, 30.3, 40.4, 50.5]
}
df1 = pd.DataFrame(data)
print("DataFrame from dictionary:")
print(df1)

# Method 2: Creating a DataFrame from a list of dictionaries
data_list = [
    {'col1': 1, 'col2': 'A', 'col3': 10.1},
    {'col1': 2, 'col2': 'B', 'col3': 20.2},
    {'col1': 3, 'col2': 'C', 'col3': 30.3},
    {'col1': 4, 'col2': 'D', 'col3': 40.4},
    {'col1': 5, 'col2': 'E', 'col3': 50.5}
]
df2 = pd.DataFrame(data_list)
print("\nDataFrame from list of dictionaries:")
print(df2)
#/////////////////////////////////////////
# Method 3: Creating a DataFrame from a NumPy array
import numpy as np
array_data = np.array([[1, 'A', 10.1],
                       [2, 'B', 20.2],
                       [3, 'C', 30.3],
                       [4, 'D', 40.4],
                       [5, 'E', 50.5]])
df3 = pd.DataFrame(array_data, columns=['col1', 'col2', 'col3']) # must supply column names
print("\nDataFrame from NumPy array:")
print(df3)

# Method 4: Creating a DataFrame from a list of lists (rows)
row_data = [[1, 'A', 10.1],
            [2, 'B', 20.2],
            [3, 'C', 30.3],
            [4, 'D', 40.4],
            [5, 'E', 50.5]]
df4 = pd.DataFrame(row_data, columns=['col1', 'col2', 'col3'])
print("\nDataFrame from list of lists:")
print(df4)

# Method 5: Creating an empty DataFrame with specified columns
empty_df = pd.DataFrame(columns=['col1', 'col2', 'col3'])
print("\nEmpty DataFrame:")
print(empty_df)

# Method 6: Creating a Dataframe from a Series.
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series(['A', 'B', 'C', 'D', 'E'])
s3 = pd.Series([10.1, 20.2, 30.3, 40.4, 50.5])

df5 = pd.DataFrame({'col1': s1, 'col2': s2, 'col3': s3})
print("\nDataFrame from Series:")
print(df5)
```
